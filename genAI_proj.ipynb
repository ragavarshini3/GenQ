```python
import google.generativeai as genai
from datetime import datetime
import random

# Configure Gemini
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# -----------------------------
# Department & Courses
# -----------------------------
DEPARTMENTS = {
    "AI&DS": {
        "name": "Artificial Intelligence and Data Science",
        "courses": {
            "Machine Learning": "Supervised Learning, Unsupervised Learning, Regression, Classification, Clustering, Feature Engineering, Model Selection",
            "Deep Learning": "Neural Networks, CNNs, RNNs, LSTMs, GANs, Transfer Learning, Activation Functions"
        }
    },
    "IT": {
        "name": "Information Technology",
        "courses": {
            "Web Development": "HTML, CSS, JavaScript, React, Node.js, REST APIs",
            "Database Management Systems": "SQL, Normalization, Indexing, ACID Properties"
        }
    }
}

# -----------------------------
# Fallback Question Templates
# -----------------------------
QUESTION_TEMPLATES = {
    "2mark": [
        "Define {topic}.",
        "What is {topic}?",
        "Write short notes on {topic}."
    ],
    "5mark": [
        "Explain {topic} with example.",
        "Discuss applications of {topic}.",
        "Describe the working of {topic}."
    ],
    "10mark": [
        "Write a detailed essay on {topic}.",
        "Analyze {topic} with real-world applications.",
        "Explain {topic} with diagram and examples."
    ]
}

# -----------------------------
# Fallback Generator
# -----------------------------
def generate_fallback(course, syllabus, two, five, ten):
    topics = [t.strip() for t in syllabus.split(",")]
    output = f"\nQUESTION PAPER - {course}\n"
    output += f"Generated on: {datetime.now()}\n\n"

    output += "SECTION A (2 Marks)\n"
    for i in range(int(two)):
        topic = random.choice(topics)
        output += f"{i+1}. {random.choice(QUESTION_TEMPLATES['2mark']).format(topic=topic)}\n"

    output += "\nSECTION B (5 Marks)\n"
    for i in range(int(five)):
        topic = random.choice(topics)
        output += f"{i+1}. {random.choice(QUESTION_TEMPLATES['5mark']).format(topic=topic)}\n"

    output += "\nSECTION C (10 Marks)\n"
    for i in range(int(ten)):
        topic = random.choice(topics)
        output += f"{i+1}. {random.choice(QUESTION_TEMPLATES['10mark']).format(topic=topic)}\n"

    return output

# -----------------------------
# AI Question Paper Generator
# -----------------------------
def generate_question_paper(department, course, difficulty, two, five, ten):
    syllabus = DEPARTMENTS[department]["courses"][course]

    prompt = f"""
Generate a university semester question paper.

Department: {DEPARTMENTS[department]['name']}
Course: {course}
Syllabus Topics: {syllabus}
Difficulty Level: {difficulty}

Create:
- {two} questions of 2 marks
- {five} questions of 5 marks
- {ten} questions of 10 marks

Format clearly with:
SECTION A
SECTION B
SECTION C

Avoid repetition and maintain academic standards.
"""

    try:
        model = genai.GenerativeModel("gemini-2.0-flash")
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        print("‚ö† API Failed. Using fallback generator.")
        return generate_fallback(course, syllabus, two, five, ten)


# -----------------------------
# üî• RUN INTERACTIVE
# -----------------------------
print("Available Departments:", list(DEPARTMENTS.keys()))
department = input("Enter Department: ")

print("Available Courses:", list(DEPARTMENTS[department]["courses"].keys()))
course = input("Enter Course: ")

difficulty = input("Enter Difficulty (Easy/Moderate/Hard): ")
two = input("Number of 2 Mark Questions: ")
five = input("Number of 5 Mark Questions: ")
ten = input("Number of 10 Mark Questions: ")

print("\n‚è≥ Generating Question Paper...\n")

paper = generate_question_paper(department, course, difficulty, two, five, ten)

print("\nüìò GENERATED QUESTION PAPER:\n")
print(paper)
```
